Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Load the Dataset

```python
df=pd.read_csv(r'C:\Users\XNovaQ.7\OneDrive - ExpertsPro\Desktop\datasets\Space_Corrected.csv', low_memory=False)
df.head()
```

Viewing the Data

```python
df.info()
df.describe(include="all")
```

The dataset used in this analysis contains detailed records of 4,324 space launches from various space agencies and private companies across the world. Each row represents a single launch event and includes information such as the launch provider, location, date and time, rocket details, mission status, and in some cases, the cost of the rocket.

Dropping Unwanted Columns 

```python
df = df.drop(columns=[c for c in df.columns if c.startswith("Unnamed")], errors="ignore")
df
```

Renaming the Columns 

```python
df = df.rename(columns={
    "Company Name": "company",
    "Location": "location",
    "Datum": "datetime_str",
    "Detail": "detail",
    "Status Rocket": "rocket_status",
    " Rocket": "rocket_cost",
    "Status Mission": "mission_status"
})
df
```

Clean Text Columns (Remove Spaces & 'nan')

```python
for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.strip().replace("nan", np.nan)
```

Convert Date Column

```python
df["datetime"] = pd.to_datetime(df["datetime_str"], errors="coerce", utc=True)
df["year"] = df["datetime"].dt.year
df["month"] = df["datetime"].dt.month
```

```python
df
```

Extracting the Country from Location info

```python
def extract_country(x):
    if pd.isna(x): return np.nan
    return x.split(",")[-1].strip()

df["country"] = df["location"].apply(extract_country)
```

```python
df
```

Clean Rocket Cost to Numeric

```python
df["rocket_cost_num"] = (
    df["rocket_cost"]
    .astype(str)
    .str.replace(r"[^0-9.]", "", regex=True)
    .replace("", np.nan)   
    .astype(float)
)
```

```python
df
```

```python
Mission Success Flag 
```

```python
df["success_flag"] = df["mission_status"].str.lower().str.contains("success", na=False)
df["success_flag"].mean()
```

```python
success_count = df["mission_status"].str.contains("Success", case=False, na=False).sum()
total_missions = len(df)
success_percent = (success_count / total_missions) * 100
print("The overall Success Percentage of Space Missions:", success_percent)
```

Launches Per Year

```python
df.groupby("year").size().plot(kind="line", figsize=(10,5))
plt.title("Launches Per Year")
plt.xlabel("Year")
plt.ylabel("Launch Count")
plt.show()
```

```python
df["mission_status"].value_counts().plot(kind="bar", figsize=(8,4))
plt.title("Mission Status Counts")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

Rocket Cost Distribution

```python
df["rocket_cost_num"].dropna().plot(kind="hist", bins=30, figsize=(8,4))
plt.title("Rocket Cost Distribution")
plt.xlabel("Cost (millions)")
plt.ylabel("Frequency")
plt.show()
```

Rocket Cost Over Years 

```python
yearly_cost = (
    df.dropna(subset=["rocket_cost_num", "year"])
      .groupby("year")["rocket_cost_num"]
      .mean()
)

# Plot
plt.figure(figsize=(12,6))
plt.plot(yearly_cost.index, yearly_cost.values, marker='o')
plt.title("Average Rocket Cost Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Rocket Cost (Millions USD)")
plt.grid(alpha=0.3)
plt.show()
```

EDA Findings: Space Launch Dataset (4,324 Records)

### 1. Mission Success Rate is Very High

89.71% of all missions were successful.

This shows strong reliability across global space programs.

Failures reduce significantly in the modern era.

### 2. Government Agencies Historically Dominated Launch Activity

From 1957–2006, almost all launches were from government agencies.

Agencies like RVSN USSR, CASC, Roscosmos, NASA/ULA were leading players.

Government launch volumes were extremely high during the Cold War and early space race.

### 3.Country-Level Insights

Countries launching the most rockets:

Russia/Kazakhstan (Baikonur legacy)

USA

China

Europe (French Guiana)

Japan & India contribute consistently

These reflect global space power distribution.

### 4. Launch Activity Trends Over Time

1960s–1980s: Extremely high launch counts driven by USA & USSR.

1990s–2000s: Decline in launch activity.

2010s–2020s: Surge again due to private players + increased satellite deployments.

### 5. Private Players Enter Late but Grow Rapidly

Private companies such as SpaceX, Rocket Lab, Blue Origin enter only after ~2006.

But from 2014 onward, their launch frequency grows sharply.

SpaceX becomes the strongest private contributor.

### 6. Cost Trend Shift After Private Sector

Before private sector:

Costs were very high (hundreds of millions to billions).

Technology improvements increased cost, not efficiency.

After private sector:

Average costs drop significantly ($40M–$90M range for many private launches).

Technology improves while cost decreases due to:

Reusability

Commercial manufacturing

Higher launch cadence

Better optimization

This is a major industry transformation.

Modern launches tend to be far cheaper (mostly < $150M).

Cost decreased while technology improved after private players entered.
