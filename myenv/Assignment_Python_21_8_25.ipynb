```python
# 2.Create a decorator that logs every time a function is called, along with its arguments, return value, and timestamp.
# Use a custom decorator called @log_to_file
# Store logs in a file named function_logs.txt
# Use functools.wraps to preserve metadata

import  functools
import datetime

def log_to_file(func):
    @functools.wraps(func)
    def wrapper(*args, **kargs):
        "This is a wrapper"
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H-%M-%S")
        result=func(*args, *kargs)

        meta_data=(
            f'''\n--------------META DATA---------------------\n
            Function Name   :{func.__name__}
            DocString       :{func.__doc__}
            Module          :{func.__doc__}
            Annotations     :{func.__annotations__}
            Defaults        :{func.__defaults__}

            \n--------------------------------------------------\n'''


        )

        log_write= (
            f'''Function: {func.__name__} was Called at [{timestamp}] 
                Arguments: {args}, kwargs={kargs} 
                Returned: {result} 
                {meta_data}'''
        )
        with open("function_logss.txt", "a") as f:
            f.write(log_write)
        return result
    return wrapper

@log_to_file
def add(a:int, b:int)->int:
    "This is the function for addition"
    return a+b

# def name(name, age=20):
#     return f"Hello {name}, Age:{age}"

print("Result Add:", add(5,4))
print("Result Add:", add(2,3))
print("Result Add:", add(1,3))
print("Result Add:", add(5,5))
print("\n")
# t=add(5,5)
#print(add.__doc__)

# with open("function_logs.txt", "r") as file:
#     content = file.read()
#     print(content)
file =open("function_logss.txt   ", "r")
content =file.read()
print(content)

# print(name("Dinesh", age=20))



```
